Zdefiniować funkcję *f(x) = x^2^ + 5* i sporządzić jej wykres dla argumentów z 3 podancyh przedziałów:  
1. x > -1 oraz x < 1
2. x > -6 oraz x < 6
3. x > 0 oraz x < 5


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x*x + 5

x1 = np.linspace(-1, 1)
y1 = f(x1)
plt.plot(x1, y1, label="f(x)")
plt.xlabel("Os X")
plt.ylabel("f(x)")
plt.title("Wykres 1")
plt.legend()
plt.show()
```

```python
x2 = np.linspace(-6, 6)
y2 = f(x2)
plt.plot(x2, y2, label="f(x)")
plt.xlabel("Os X")
plt.ylabel("f(x)")
plt.title("Wykres 2")
plt.legend()
plt.show()
```

```python
x3 = np.linspace(0, 5)
y3 = f(x3)
plt.plot(x3, y3, label="f(x)")
plt.xlabel("Os X")
plt.ylabel("f(x)")
plt.title("Wykres 3")
plt.legend()
plt.show()
```
